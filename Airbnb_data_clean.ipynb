<h2>Cleaning individual columns</h2>

```python
df
```

```python
# 'Host_identity_verified' --> formatting to upper()
df['Host_identity_verified']=df['Host_identity_verified'].str.upper()
```

```python
# 'Cancellation_policy' --> formatting to capitallize()
df['Cancellation_policy']=df['Cancellation_policy'].str.capitalize()
```

```python
df
```

```python
# 'Construction year' --> removing '.0' in the year ie (2022.0)
# 1. change the datatype to str to perform replace()

df['Construction year']=df['Construction year'].astype(str)
```

```python
df.dtypes
```

```python
df['Construction year']=df['Construction year'].str.replace('.0','',regex=False)
df
```

```python
# 'Price & Service fee' --> replacing '$,' by blank space, to perform numeric operations 

df['Price']=df['Price'].str.replace(',','',regex=False)
df['Price']=df['Price'].str.replace('$','',regex=False)

df['Service fee']=df['Service fee'].str.replace('[$,]','',regex=True)
df
# other way by giving the character in list & assigning 'regex=True'
# df['Price']=df['Price'].str.replace('[$,]','',regex=True)
```

```python
df['Service fee']
```

```python
df.info()
```

```python
# converted float to int datatype
df[['Minimum nights','Number of reviews']]=df[['Minimum nights','Number of reviews']].astype(int)
df
```

```python
# converted True/False to 1-0's
df['Instant_bookable']=df['Instant_bookable'].apply(lambda x: 1 if x==True else 0)
df
```

```python
# converting the datatype to 'int'
df[['Instant_bookable','Construction year','Price','Service fee','Minimum nights','Number of reviews']]=df[['Instant_bookable','Construction year','Price','Service fee','Minimum nights','Number of reviews']].astype('int64')
```

```python
df.info()
```

```python
df.reset_index(inplace=True)
df
```

```python
df.drop(columns='index',inplace=True)
df
```

```python
df
```

```python
# replacing with '_' instead of blank space
df.columns=df.columns.str.replace(' ','_')
df
```

```python
# saving the data to .csv
df.to_csv(r"Cleaned_Airbnb_dataset.csv")
```
